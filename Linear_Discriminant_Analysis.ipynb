# Linear Discriminant Analysis

Linear Discriminant Analysis (LDA) is classification algorithm that finds a linear combination of features to best separate different classes in a dataset.

Types of Linear Discriminant Analysis :

1. Binary LDA : two classes
2. Multiclass LDA : for separating more than two classes
based on linear decision boundaries

-> Following Project is an exmaple of Binary LDA

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

fruit_features = np.array([[3,7],[2,8],[3,6],[4,7],[1,4],[2,3]]) #[size , sweetness]

fruit_labels = np.array([1,1,1,1,0,0])

model=LinearDiscriminantAnalysis()

model.fit(fruit_features,fruit_labels)

predict_size = float(input("Enter the size of the fruit : "))

predict_sweetness = float(input("Enter the sweetness of the fruit : "))

new_fruit=np.array([[predict_size,predict_sweetness]])

predicted_label=model.predict(new_fruit)

print(f"User {'likes' if predicted_label[0]==1 else 'notlike'} a fruit of size {new_fruit[0,0]} and sweetness {new_fruit[0,1]}")
```

```python
np.random.seed(42)

sample_class_0 = np.random.rand(20, 2) * 5 + [1, 2]

sample_class_1 = np.random.rand(20, 2) * 5 + [6, 7]


plt.scatter(fruit_features[:,0], fruit_features[:,1], c=fruit_labels, cmap='viridis', marker='o', label='Original Data')

plt.scatter(new_fruit[:,0], new_fruit[:,1], color='darkred', marker='x', s=100, label='Predicted Fruit')

plt.scatter(sample_class_0[:,0], sample_class_0[:,1], color='blue', alpha=0.5, label='Sample Class 0')

plt.scatter(sample_class_1[:,0], sample_class_1[:,1], color='green', alpha=0.5, label='Sample Class 1')

plt.xlabel('Size')

plt.ylabel('Sweetness')

plt.title("Linear Discriminant Analysis with Sample Data")

plt.legend(prop = {'size' : 8} )

plt.grid(True)

plt.show()
```
