# Decision Tree Algorithm

A decision tree is a supervised learning algorithm that makes decisions by splitting data into branches based on feature values, forming a tree-like structure for classification or prediction

* Decision Tree Classifier: It is used to predict categories (such as yes/no or spam/not spam) by splitting data based on feature conditions.

* Decision Tree Regressor: It is used to predict continuous numerical values (such as price or temperature) by splitting data to minimize prediction error.

-> Following project uses DecisionTreeRegressor

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor, plot_tree

study_hours = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)

score_obtained = np.array([50, 55, 70, 80, 85, 90, 92, 98])

model = DecisionTreeRegressor(max_depth=3)

model.fit(study_hours, score_obtained)

predict_time = float(input("Enter the time studied by user: "))

predicted_score = model.predict([[predict_time]])

print(f"Predicted test score for {predict_time} hours: {predicted_score[0]:.2f}")
```

```python
plt.figure(figsize=(12,8))

plot_tree(model , filled=True , rounded = True , feature_names = ['study_hours'])

plt.title('Decision Tree Algorithm ')

plt.show()
```

```python
plt.scatter(study_hours,score_obtained,color='darkred')

plt.plot(study_hours,model.predict(study_hours),color='orange')

plt.scatter(predict_time,predicted_score,color='green')

plt.xlabel('Study Hours')

plt.ylabel('Score Obtained')

plt.title('Decision Tree Algorithm')

plt.show()
```
