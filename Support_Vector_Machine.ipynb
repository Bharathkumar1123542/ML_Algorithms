# Support_Vector_Machine

Support Vector Machine (SVM) is a supervised learning algorithm that finds the optimal boundary (hyperplane) to separate data points of different classes with maximum margin

* Support Vector Classifier (SVC): It separates data into different classes by finding the best hyperplane with maximum margin.

* Support Vector Regressor (SVR): It predicts continuous values by fitting a function within a specified error margin.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

X = np.array([
    [2, 3],
    [3, 4],
    [4, 5],
    [6, 7],
    [7, 8],
    [8, 9]
])

y = np.array([0, 0, 0, 1, 1, 1])

model = SVC(kernel='linear', C=1.0)

model.fit(X, y)

w = model.coef_[0]

b = model.intercept_[0]
```

```python
slope = -w[0] / w[1]

intercept = -b / w[1]

x_points = np.linspace(1, 9, 100)

y_points = slope * x_points + intercept

margin = 1 / w[1]
y_margin_upper = y_points + margin
y_margin_lower = y_points - margin

plt.figure(figsize=(8, 6))

plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', s=100, edgecolors='k', label="Data Points")

plt.plot(x_points, y_points, 'k-', label="Decision Boundary")

plt.plot(x_points, y_margin_upper, 'k--', alpha=0.5, label="Margins")

plt.plot(x_points, y_margin_lower, 'k--', alpha=0.5)

plt.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1],
            s=250, facecolors='none', edgecolors='g', linewidths=2, label="Support Vectors")

plt.xlabel("Feature 1")

plt.ylabel("Feature 2")

plt.title("SVM Linear Classification")

plt.legend()

plt.grid(True)

plt.show()
```
