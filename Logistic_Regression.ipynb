# Logistic Regression

Logistic regression is an classification problems, where the output is a probability that is mapped to categories such as yes/no or true/false.

types of logistic regression
1. binary logistic regression : two possible outcomes
2. multinomial logistic regression : for more than two categories

-> Following Project is an example of Binary Logistic Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

pages = np.array([100,150,200,250,300,350,400,450,500]).reshape(-1,1)

likes = np.array([0,1,1,1,0,0,0,0,0])

model = LogisticRegression()

model.fit(pages, likes)

predict_book_pages = int(input("Enter the Number of pages in a book : "))

predicted_result = model.predict([[predict_book_pages]])

print(f"User will {'like' if predicted_result[0] == 1 else 'notlike'} a book {predict_book_pages} pages")
```

```python
plt.scatter(pages , likes , color = 'forestgreen')

plt.plot(pages , model.predict_proba(pages)[:,1] , color = 'darkred')

plt.xlabel('Number of Pages')

plt.ylabel('Likelihood of liking')

plt.title('Logistic Regression')

plt.axvline(x=predict_book_pages,color='green',linestyle='--')

plt.axhline(y=0.5,color='red',linestyle='--')

plt.show()
```
